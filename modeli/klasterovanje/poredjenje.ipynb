```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from joblib import load
```

```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
```

```python
X, y = load('../../preprocesiranje/clustering_data.joblib')
```

```python
X.drop(['age', 'gender', 'travel_time', 'dislike_hw'], inplace = True, axis = 1)
X
```

```python
time_change = X['time_dp'] - X['time_bp']
X['time_change'] = time_change

X.drop(['time_dp', 'time_bp'], inplace = True, axis = 1)
X
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(data = X_scaled, columns = X.columns)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 3)

X_pca = pca.fit_transform(X_scaled)
```

```python
from sklearn.metrics import silhouette_score


silhouette_scores_kmeans = []
silhouette_scores_aggl = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters = k, n_init = 12, random_state = 42)
    kmeans.fit(X_pca)

    aggl = AgglomerativeClustering(n_clusters = k, linkage = 'average')
    aggl.fit(X_pca)
    
    silhouette_score_k = silhouette_score(X_pca, kmeans.labels_)
    silhouette_scores_kmeans.append(silhouette_score_k)

    silhouette_score_a = silhouette_score(X_pca, aggl.labels_)
    silhouette_scores_aggl.append(silhouette_score_a)


# Plot Silhouette Scores
plt.figure(figsize=(8, 4))
plt.plot(range(2, 11), silhouette_scores_kmeans, label='KMeans', marker='o')
plt.plot(range(2, 11), silhouette_scores_aggl, label='Agglomerative', marker='o')

plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal k')
plt.legend()

plt.tight_layout()
plt.show()
```

Na osnovu analize silhouette skorova, možemo zaključiti da je model K-Means verovatno bolji od modela Agglomerative Clustering za klasterizaciju ovih podataka.
